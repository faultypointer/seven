# DataFrame

## Creating DF from dict

```python
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Math' : [76, 68, 87, 56, 35],
    'English' : [69, 87, 81, 66, 59],
    'Science' : [92, 65, 78, 46, 66],
}
df = pd.DataFrame(data)
df
```

## Viewing data in dataframe

```python
df.head(2)
```

```python
df.tail(3)
```

```python
df.sample(2, random_state=43)
```

```python
df.info()
```

## Adding a new column

```python
df['Total'] = df[['Math', 'English', 'Science']].sum(axis=1)
```

```python
df
```

## Dropping a column

```python
df_no_ingles = df.drop(columns=['English'])
df_no_ingles
```

```python
df
```

## Renaming a column

```python
df_no_ingles.rename(columns={'Science': 'Ciencia'}, inplace=True)
df_no_ingles
```

## Missing Value

## Filling missing values in numeric column

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
df
```
